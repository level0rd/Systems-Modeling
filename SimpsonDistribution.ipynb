```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 500
c = (a+b)/2
x_pl = []
h_pl = []
total = 0
    
while total<500:
    x = np.random.uniform(a,b)
    x_pl.append(x)
    total += 1
    if (x >= a) and (x <= c):
        h_pl.append(2*(x-a)/((b-a)*(c-a)))
    elif (x>c and x<=b): 
        h_pl.append(2*(b-x)/((b-a)*(b-c)))
    else:
        h_pl.append(0)
    
    
plt.figure(figsize=[8,8])  
plt.title('График плотности треугольного распределения случайной величины') 
plt.scatter(x_pl, h_pl, color='r', s=1)
plt.show()  
```

```python
import numpy as np

a = 0
b = 500
c = (a+b)/2
x_pl = []
h_pl = []
total = 0
    
while total<500:
    x = np.random.uniform(a,b)
    x_pl.append(x)
    total += 1
    if (x >= a and x <= c):
        h_pl.append((x-a)**2/((b-a)*(c-a)))
    elif (x>c and x<=b): 
        h_pl.append(1-(b-x)**2/((b-a)*(b-c)))
    else:
        h_pl.append(0)
    
    
plt.figure(figsize=[8,8]) 
plt.title('График функции треугольного распределения') 
plt.scatter(x_pl, h_pl, color='r', s=1)
plt.show() 
```
