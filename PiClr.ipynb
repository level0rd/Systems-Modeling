```python
import math
import matplotlib.pyplot as plt
import numpy as np

x_in = []
y_in = []
x_out = []
y_out = []

r = 1
total = 0
circle = 0

while total<10000:
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    total += 1
    
    d=math.sqrt(x*x+y*y)
    if d<r:
        x_in.append(x) 
        y_in.append(y)
        circle += 1
    else:
        x_out.append(x)
        y_out.append(y)
        

pi = 4*(circle/total)
print(pi) 

plt.figure(figsize=[5,5])  
plt.scatter(x_out,y_out, s=1)
plt.scatter(x_in, y_in, color='r', s=1)
plt.show()              
```
