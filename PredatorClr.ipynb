```python
import matplotlib.pyplot as plt
import numpy as np

a = 1.1 
b = 0.9
c = 0.5
d = 1.0
xpoints, ypoints  = [], []
 
def rk4(r, t, h):
    k1 = h*f(r, t)
    k2 = h*f(r+0.5*k1, t+0.5*h)
    k3 = h*f(r+0.5*k2, t+0.5*h)
    k4 = h*f(r+k3, t+h)
    return (k1 + 2*k2 + 2*k3 + k4)
 
def f(r, t):
    x, y = r[0], r[1]
    fxd = x*(a - b*y)
    fyd = -y*(c - d*x)
    return np.array([fxd, fyd], float)
 
h=0.001                               
tpoints = np.arange(0, 10, h)         
r = np.array([2, 2], float)
for t in tpoints:
        xpoints.append(r[0])         
        ypoints.append(r[1])          
        r += rk4(r, t, h)             
            
plt.plot(tpoints, ypoints)
plt.xlabel("Время")
plt.ylabel("Популяция")
plt.show()
```
